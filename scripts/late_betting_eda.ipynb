```python

import requests
import json
```

```python
a="""Host: www.aspiturf.com:1338
User-Agent: Mozilla/5.0 (X11; Linux x86_64; rv:91.0) Gecko/20100101 Firefox/91.0
Accept: */*
Accept-Language: en-US,en;q=0.5
Accept-Encoding: gzip, deflate, br
Content-Type: text/plain
Content-Length: 261
Origin: https://aspiturf.com
DNT: 1
Connection: keep-alive
Referer: https://aspiturf.com/
Sec-Fetch-Dest: empty
Sec-Fetch-Mode: cors
Sec-Fetch-Site: same-site
Cache-Control: max-age=0"""
headers={l.split(': ')[0]:l.split(': ')[1] for l in a.splitlines()}
```

```python
query_url='https://www.aspiturf.com:1338/parse/classes/BetFormated'
response = requests.post(query_url, 
                        headers=headers,
                        data=json.dumps({"where":{"jour":"2021-05-29","reun":"1","prix":"1"},
                              "limit":2000,"order":"-updatedAt","_method":"GET","_ApplicationId":"APPLICATION_ID",
                              "_JavaScriptKey":"YOUR_JAVASCRIPT_KEY",
                              "_ClientVersion":"js2.19.0","_InstallationId":"c090473d-c969-4ebe-8ec2-9e1eb29f2cf7"}),)

response.raise_for_status()
response.json()
```

```python
for result in response.json()['results']:
    print(result['createdAt'], result['updatedAt'], result['isDepartImminent'], sum(result['enjeuInlineSG'].values()))
```
